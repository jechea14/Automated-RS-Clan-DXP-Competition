```python
import csv
import pandas as pd

CSV1 = 'clanmate_data1_11_05_2022.csv'
CSV2 = 'clanmate_data1_11_06_2022.csv'
```

```python
file1_df = pd.read_csv(CSV1)
file2_df = pd.read_csv(CSV2)

file1_df
```

```python
df3 = file2_df[file2_df.columns[3:]] - file1_df[file1_df.columns[3:]]
df3['Name'] = file2_df['Name']
df3['Total Lvl'] = file2_df['Total Lvl']
cols = list(df3.columns.values) 
df3 = df3[[cols[-2]] + [cols[-1]] + cols[0:28]]
df3
```

```python
halved_skills = [
    'Attack',
    'Defense',
    'Strength',
    'Constitution',
    'Ranged',
    'Magic',
    'Summoning',
    'Herblore',
    'Farming',
    'Invention',
    'Archaeology',
    'Dungeoneering'
]

normal_skills = [
    'Fletching',
    'Crafting',
    'Thieving'
]

doubled_skills = [
    'Hunter',
    'Smithing',
    'Firemaking',
    'Prayer',
    'Cooking',
    'Slayer',
    'Construction'
]

tripled_skills = [
    'Agility',
    'Divination',
    'Fishing',
    'Woodcutting',
    'Mining',
    'Runecrafting'
]
calculated_data = {}
for halved in halved_skills:
    calculated_data[halved] = df3[halved] * 0.5
    
for normal in normal_skills:
    calculated_data[normal] = df3[normal] * 1

for doubled in doubled_skills:
    calculated_data[doubled] = df3[doubled] * 2

for tripled in tripled_skills:
    calculated_data[tripled] = df3[tripled] * 3

calculated_df = pd.DataFrame(data=calculated_data)
calculated_df['Name'] = df3['Name']
calculated_df['Total Lvl'] = df3['Total Lvl']
calculated_df['Total'] = calculated_df.loc[:, 'Attack':'Runecrafting'].sum(axis=1)
cols = list(calculated_df.columns.values) 
calculated_df = calculated_df[[cols[-3]] + [cols[-2]] + [cols[-1]] + cols[0:28]]

calculated_df
```
