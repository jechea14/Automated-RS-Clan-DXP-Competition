```python
import csv
import pandas as pd
from datetime import datetime

CSV1 = 'clanmate_data1_11_05_2022.csv'
CSV2 = 'clanmate_data1_11_06_2022.csv'
```

```python
file1_df = pd.read_csv(CSV1)
file2_df = pd.read_csv(CSV2)

file1_df
```

```python
df3 = file2_df[file2_df.columns[3:]] - file1_df[file1_df.columns[3:]]
df3['Name'] = file1_df['Name']
df3['Total Lvl'] = file1_df['Total Lvl']
cols = list(df3.columns.values) 
df3 = df3[[cols[-2]] + [cols[-1]] + cols[0:28]]
df3
```

```python
halved_skills = [
    'Attack',
    'Defense',
    'Strength',
    'Constitution',
    'Ranged',
    'Magic',
    'Summoning',
    'Herblore',
    'Farming',
    'Invention',
    'Archaeology',
    'Dungeoneering'
]

normal_skills = [
    'Fletching',
    'Crafting',
    'Thieving'
]

doubled_skills = [
    'Hunter',
    'Smithing',
    'Firemaking',
    'Prayer',
    'Cooking',
    'Slayer',
    'Construction'
]

tripled_skills = [
    'Agility',
    'Divination',
    'Fishing',
    'Woodcutting',
    'Mining',
    'Runecrafting'
]
calculated_data = {}
for halved in halved_skills:
    calculated_data[halved] = df3[halved] * 0.5
    
for normal in normal_skills:
    calculated_data[normal] = df3[normal] * 1

for doubled in doubled_skills:
    calculated_data[doubled] = df3[doubled] * 2

for tripled in tripled_skills:
    calculated_data[tripled] = df3[tripled] * 3

calculated_df = pd.DataFrame(data=calculated_data)
calculated_df['Name'] = df3['Name']
calculated_df['Total Lvl'] = df3['Total Lvl']
calculated_df['DXP Gained'] = calculated_df.loc[:, 'Attack':'Runecrafting'].sum(axis=1)
cols = list(calculated_df.columns.values) 
calculated_df = calculated_df[[cols[-3]] + [cols[-1]] + [cols[-2]] + cols[0:28]]

calculated_df
```

```python
bracketG_data = calculated_df.loc[(calculated_df['Total Lvl'] >= 2851)].sort_values('DXP Gained', ascending=False)
bracketG_data['Rank'] = range(1, len(bracketG_data) + 1)
first_column = bracketG_data.pop('Rank')
# bracketG_data[['Rank', 'Name', 'DXP Gained']].loc[bracketG_data['DXP Gained'] > 0]
bracketG_data.insert(0, 'Rank', first_column)
bracketG_data
```

```python
bracketA_data = calculated_df.loc[(calculated_df['Total Lvl'] <= 2000)].sort_values('DXP Gained', ascending=False)
bracketA_data['Rank'] = range(1, len(bracketA_data) + 1)

# bracketA_data[['Rank', 'Name', 'DXP Gained']].loc[bracketA_data['DXP Gained'] > 0]
first_column = bracketA_data.pop('Rank')
bracketA_data.insert(0, 'Rank', first_column)
bracketA_data
```

```python
bracketB_data = calculated_df.loc[(calculated_df['Total Lvl'] >= 2001) & (calculated_df['Total Lvl'] <= 2300)].sort_values('DXP Gained', ascending=False)
bracketB_data['Rank'] = range(1, len(bracketB_data) + 1)

# bracketB_data[['Rank', 'Name', 'DXP Gained']].loc[bracketB_data['DXP Gained'] > 0]
first_column = bracketB_data.pop('Rank')
bracketB_data.insert(0, 'Rank', first_column)
bracketB_data
```

```python
bracketC_data = calculated_df.loc[(calculated_df['Total Lvl'] >= 2301) & (calculated_df['Total Lvl'] <= 2600)].sort_values('DXP Gained', ascending=False)
bracketC_data['Rank'] = range(1, len(bracketC_data) + 1)

# bracketC_data[['Rank', 'Name', 'DXP Gained']].loc[bracketC_data['DXP Gained'] > 0]
first_column = bracketC_data.pop('Rank')
bracketC_data.insert(0, 'Rank', first_column)
bracketC_data
```

```python
bracketD_data = calculated_df.loc[(calculated_df['Total Lvl'] >= 2601) & (calculated_df['Total Lvl'] <= 2700)].sort_values('DXP Gained', ascending=False)
bracketD_data['Rank'] = range(1, len(bracketD_data) + 1)

# bracketD_data[['Rank', 'Name', 'DXP Gained']].loc[bracketD_data['DXP Gained'] > 0]
first_column = bracketD_data.pop('Rank')
bracketD_data.insert(0, 'Rank', first_column)
bracketD_data
```

```python
bracketE_data = calculated_df.loc[(calculated_df['Total Lvl'] >= 2701) & (calculated_df['Total Lvl'] <= 2800)].sort_values('DXP Gained', ascending=False)
bracketE_data['Rank'] = range(1, len(bracketE_data) + 1)

# bracketE_data[['Rank', 'Name', 'DXP Gained']].loc[bracketE_data['DXP Gained'] > 0]
first_column = bracketE_data.pop('Rank')
bracketE_data.insert(0, 'Rank', first_column)
bracketE_data
```

```python
bracketF_data = calculated_df.loc[(calculated_df['Total Lvl'] >= 2801) & (calculated_df['Total Lvl'] <= 2850)].sort_values('DXP Gained', ascending=False)
bracketF_data['Rank'] = range(1, len(bracketF_data) + 1)

# bracketF_data[['Rank', 'Name', 'DXP Gained']].loc[bracketF_data['DXP Gained'] > 0]
first_column = bracketF_data.pop('Rank')
bracketF_data.insert(0, 'Rank', first_column)
bracketF_data
```

```python

date = datetime.now().strftime("%m_%d_%Y")
date_string = "results_" + date + ".xlsx"
with pd.ExcelWriter(date_string) as writer:
    bracketA_data.to_excel(writer, sheet_name="Bracket A < 2k Total")
    bracketB_data.to_excel(writer, sheet_name="Bracket B 2k - 2.3k Total")
    bracketC_data.to_excel(
        writer, sheet_name="Bracket C 2.3k - 2.6k Total")
    bracketD_data.to_excel(
        writer, sheet_name="Bracket D 2.6k - 2.7k Total")
    bracketE_data.to_excel(
        writer, sheet_name="Bracket E 2.7k - 2.8k Total")
    bracketF_data.to_excel(
        writer, sheet_name="Bracket F 2.8k - 2850 Total")
    bracketG_data.to_excel(writer, sheet_name="Bracket G > 2850 Total")
```
